```python
#Setting the aesthetic style of the plots.
sns.set_style("whitegrid")
```

```python
# Population trends over the years.
years = ['1970 population', '1980 population', '1990 population', '2000 population', '2010 population', '2015 population', '2020 population', '2022 population', '2023 population']
population_trends = wld_data[years].sum()
```

```python
#Population Distribution by Continent in 2023.
continent_population = wld_data.groupby('continent')['2023 population'].sum().sort_values(ascending=False)
```

```python
# Population, Density and Area Relationship.
plt.figure(figsize=(10, 6))
sns.scatterplot(data=wld_data, x='area (kmÂ²)', y='density (kmÂ²)')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Area (km²) [Log Scale]')
plt.ylabel('Density (km²) [Log Scale]')
plt.title('Population Density vs Area of Countries (Log Scale)')
```

```python
# Plotting the results
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
population_trends.plot(kind='bar', color='blue')
plt.title('Global Population Trends (1970 - 2023)')
plt.ylabel('Total Population')
plt.xlabel('Year')
```

```python
# Population distribution in terms of continent.
plt.subplot(1, 2, 2)
continent_population.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Population Distribution by Continent (2023)')
plt.ylabel('')

plt.tight_layout()
plt.show()
```
